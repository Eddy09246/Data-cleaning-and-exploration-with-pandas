## Drop Rows with NaN

```python
clean_df = df.dropna()
```

## Accessing columns in dataframe

```python
clean_df['Starting Median Salary']
clean_df['Starting Median Salary'].max()
clean_df["Starting Median Salary"].idxmax()
clean_df['Undergraduate Major'].loc[43]
clean_df['Undergraduate Major'][43]
```

Sorting Values & Adding Columns: Majors with the Most Potential vs Lowest Risk

```python
pencentile_diff=clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', pencentile_diff)
clean_df.head()
```

Sorting by the Lowest Spread

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

## Sorting by the Highest Mid-Career 90th Percentile Salary

```python
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```
